```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag, ne_chunk

# Download required NLTK data (run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

class NLTKProcessor:
    def __init__(self):
        self.stemmer = PorterStemmer()
        self.lemmatizer = WordNetLemmatizer()
        self.stop_words = set(stopwords.words('english'))
        
    def process_text(self, text):
        print("=" * 50)
        print("NLTK Text Processing")
        print("=" * 50)
        
        # 1. Sentence Tokenization
        sentences = sent_tokenize(text)
        print("\n1. Sentences:")
        for i, sent in enumerate(sentences, 1):
            print(f"   Sentence {i}: {sent}")
        
        # 2. Word Tokenization
        words = word_tokenize(text)
        print(f"\n2. Word Tokens ({len(words)} words):")
        print(f"   {words}")
        
        # 3. Stop Words Removal
        filtered_words = [w for w in words if w.lower() not in self.stop_words and w.isalpha()]
        print(f"\n3. After Stop Words Removal ({len(filtered_words)} words):")
        print(f"   {filtered_words}")
        
        # 4. Stemming
        stemmed = [self.stemmer.stem(w) for w in filtered_words]
        print(f"\n4. Stemmed Words:")
        print(f"   {stemmed}")
        
        # 5. Lemmatization
        lemmatized = [self.lemmatizer.lemmatize(w) for w in filtered_words]
        print(f"\n5. Lemmatized Words:")
        print(f"   {lemmatized}")
        
        # 6. Part-of-Speech Tagging
        pos_tags = pos_tag(words)
        print(f"\n6. POS Tagging (first 10):")
        for word, tag in pos_tags[:10]:
            print(f"   {word} -> {tag}")
        
        # 7. Named Entity Recognition (simplified)
        chunked = ne_chunk(pos_tags)
        print(f"\n7. Named Entities:")
        entities = []
        for chunk in chunked:
            if hasattr(chunk, 'label'):
                entities.append(f"{' '.join(c[0] for c in chunk)} ({chunk.label()})")
        if entities:
            for entity in entities:
                print(f"   {entity}")
        else:
            print("   No named entities found")
        
        return {
            'sentences': sentences,
            'tokens': words,
            'filtered': filtered_words,
            'stemmed': stemmed,
            'lemmatized': lemmatized,
            'pos_tags': pos_tags
        }

# Example usage
if __name__ == "__main__":
    sample_text = """
    Natural Language Processing with NLTK is fascinating. 
    Dr. John Smith works at Stanford University in California. 
    He is developing advanced algorithms for text analysis.
    The project was funded by Google and Microsoft in 2023.
    """
    
    processor = NLTKProcessor()
    results = processor.process_text(sample_text)
```
